## Parcoursup 2019 - vœux de poursuite d'études et de réorientation dans l'enseignement supérieur et réponses des établissements 

#### Date de dernière modification : 27 avril 2020 11:22

#### Producteur : Ministère de l'Enseignement supérieur, de la Recherche et de l'Innovation > Sous-direction des Systèmes d'information et études statistiques

#### Langue : Français

##### Explication : https://data.enseignementsup-recherche.gouv.fr/explore/dataset/fr-esr-parcoursup/information/?sort=tri

##### Download .csv : https://data.enseignementsup-recherche.gouv.fr/explore/dataset/fr-esr-parcoursup/download/?format=csv&timezone=Europe/Berlin&lang=fr&use_labels_for_header=true&csv_separator=%3B

Ce jeu de données présente les vœux de poursuite d’études et de réorientation dans l’enseignement supérieur ainsi que les propositions des établissements pour chaque formation – hors apprentissage – à la fin du processus d’affectation de la plateforme Parcoursup pour la session 2019 (du 22 janvier au 14 septembre 2019). Les données sont tirées de la plateforme Parcoursup 2019 et prennent en compte l’ensemble de la campagne jusqu’à la date du 14 septembre 2019.

Il couvre l’ensemble des candidats ayant au moins un vœu d’orientation validé en phase principale et/ou complémentaire, ce parmi les 11 588 formations proposées hors apprentissage. Il couvre ainsi 937 332 candidats. Un sous ensemble de données porte spécifiquement sur les 652 865 néo-bacheliers parmi ces candidats.

### Les informations sont accessibles selon l’établissement d’accueil et la formation :

Le nom de l’établissement et son n°UAI

La région, l’académie et le département de l’établissement

La formation demandée, agrégée et détaillée

Le nombre de places (capacité d’accueil)

Le nombre de vœux reçus selon le sexe et la série du baccalauréat

Le nombre de vœux classés par l’établissement de la formation d’accueil

Le nombre de propositions faites

Le nombre de propositions faites aux différentes dates clés de la campagne et qui seront ensuite acceptées

Le nombre d’admis selon le sexe, la série du baccalauréat et la mention obtenue

Le rang du dernier appelé parmi les candidatures en phase principale

Le nombre de boursiers du secondaire pour les terminales et de l’enseignement supérieur pour les étudiants en réorientation

Le nombre de candidats admis en BTS ou en CPGE et issus de leur lycée (i.e. le lycée où ils étaient inscrits en terminale)

Il permet ainsi d’observer pour chaque formation présente dans Parcoursup et pour chaque établissement d’accueil la demande et la satisfaction de celle-ci ; le profil des candidats et celui des admis ainsi que d’autres résultats à la fin du processus d’affectation.

```python
#Creation of variable to facilitate column call
statut = "Statut de l’établissement de la filière de formation (public, privé…)"

ef_total_candi = "Effectif total des candidats pour une formation"                                                              

etab = "Établissement"

ef_total_accep = "Effectif total des candidats ayant accepté la proposition de l’établissement (admis)"

cap_etab = "Capacité de l’établissement par formation"

filiere = "Filière de formation"
```

## Distribution of private and public formations on Parcoursup


```python
# Double [[]] to display a table
parcoursup.groupby(statut).count()[[etab]].sort_values(by=etab,ascending=False)
#Répartition des formations privées et publiques sur Parcoursup
#Distribution of private and public formations on Parcoursup
```

Doc sort_values : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html

## Number of candidates accepted for public or private training

```python
parcoursup.groupby(statut).sum()[[ef_total_accep]].sort_values(by=ef_total_accep,ascending=False)
#Nombre de candidats admis à une formation publique ou privée
#Number of candidates accepted for public or private training
```

```python
parcoursup[(parcoursup[filiere]=="Ecoles de commerce et de management")].sort_values(by=ef_total_candi,ascending = False)[[etab,ef_total_candi]]
```

### List of engineering training courses compared to the number of candidates on Parcoursup

```python
#Liste des formations d'ingénieurs par rapport aux nombres de candidats sur Parcoursup
#List of engineering training courses compared to the number of candidates on Parcoursup
inge_list = parcoursup[(parcoursup[filiere]=="Formations d'ingénieurs")].sort_values(by=ef_total_candi,ascending = False)[[etab,ef_total_candi]]
inge_list
```

### Rankings of the most highly ranked engineering schools on Parcourpsup

```python
#Classement des écoles d'ingénieurs les plus candidatées sur Parcourpsup
#Rankings of the most highly ranked engineering schools on Parcourpsup
inge_list.groupby(etab).sum()[[ef_total_candi]].sort_values(by=ef_total_candi,ascending=False)
```

```python
ef_total_candi_statut = parcoursup.groupby(statut).sum()[cap_etab]
#Histogramme de la capacitité en nombre d'étudiant des formations privées et publiques avec Seaborn
#Histogram of capacity in number of students in private and public training courses with Seaborn
sns.barplot(y=ef_total_candi_statut.index,x=ef_total_candi_statut.values).set_title("Histogram of the reappearance of private and public training with Seaborn")
#Seaborn's barplot returns an axis-object (not a figure).
plt.xlabel("Nombres de places disponibles pour une formation")
plt.ylabel("Statut de l'établisement")
#plt.title("Title")
plt.show()
```

### Number of engineering schools according to the entrance examination

### Number of engineering schools according to the entrance examination with a scientific baccalaureate"
