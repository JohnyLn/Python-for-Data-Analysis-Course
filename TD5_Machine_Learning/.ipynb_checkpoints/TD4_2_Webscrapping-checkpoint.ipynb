```python
css_selector = "img"
images       = driver.find_elements_by_css_selector(css_selector)
img_srcset   = [image.get_attribute('srcset') for image in images]
img_srcset
```

```python

# We're filtering out all strings that don't start with "http"
img_srcsetclean = list(filter(lambda url: url.startswith("http"), img_srcset))
img_srcsetclean
```

```python
urls_images  = [x.split(" ")[4] for x in img_srcsetclean]
urls_images
#[url for url in urls_images if re.search(r"5FC4A050",url)]
```
