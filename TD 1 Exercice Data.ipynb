Average price in UK House Price Index January 2020

```python
data = pd.read_csv("Average-prices-2020-01.csv",sep=",")
```

```python
data.groupby("Region_Name").Average_Price.mean()
```

```python
york_data = data[(data.Region_Name=="York")].sort_values(by="Date")
```

```python
york_data.head()
```

```python
plt.plot(york_data.Date,york_data.Average_Price)
```

```python
plt.rcParams['figure.figsize'] = [20,8]
```

```python
plt.rcParams['figure.dpi'] = 200
```

```python
plt.xlabel("Date")
plt.ylabel("Price Average")
plt.plot(title="The price Average in UK House between 1995 and 2020")
plt.plot(york_data.Date,york_data.Average_Price)
```
